```python
addr = pd.read_csv('address.csv', index_col=0, low_memory=False)
addr.head(1)
```

```python
cols_to_drop = [
    'licence',
    'display_name',
    'address',
    'boundingbox',
    'ISO3166-2-lvl4',
    'state',
    'country',
    'country_code'
      ]
addr = addr.drop(columns=cols_to_drop)
addr.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()

temp = addr[~addr.city.isna()]
x = temp[['lat','lon']]
y = temp.city

knc.fit(x, y)
p = knc.predict(x)
(p == y).mean()

addr[addr.city.isna()].city = knc.predict(addr[addr.city.isna()][['lat','lon']])
```

```python
addr[addr.city.isna()]
```

```python
amask = addr.hamlet.isna()
bmask = addr.city.isna()
cmask = addr.suburb.isna()
dmask = addr.neighbourhood.isna()
emask = addr.village.isna()
addr[(amask & bmask & cmask & dmask & emask)].shape
```

```python
import plotly.express as px
fig = px.scatter_mapbox(addr[(amask & bmask & cmask & dmask & emask)], lat="lat", lon="lon",
                    # color="fips", 
                    opacity=.8,
                    size_max=.5, zoom=7)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":20,"t":0,"l":0,"b":0})
fig.update_layout(width=800, height= 700)
fig.show()
```
