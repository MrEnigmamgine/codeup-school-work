1. Use matplotlib to plot the following equation:

$y = x^2 - x + 2$

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(-49,50))
y = [n**2 - n + 2 for n in x]

print('x =', x)
print('y =', y)
```

```python
plt.plot(x,y)
plt.scatter(0,0)
plt.annotate('Origin', 
              xy=(0,0) ,
              xytext=(-5,300),
              arrowprops={
                'arrowstyle':'->'  
              }
            )
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
You can use functions from the math module to help implement some of the equations above.

$y = \sqrt{x}$

$y = x^3$

$y = 2^x$

$y = 1/(x+1)$

```python
import math

def chart1(n1, n2): 
    # Chart for y = \sqrt{x}
    x = range(n1,n2)
    y = [math.sqrt(n) for n in x]
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('$\sqrt{x}$')
    # plt.show()

def chart2(n1, n2):
    # Chart for y = x^3
    x = range(n1,n2)
    y = [n**3 for n in x]
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('${x}^3$')
    # plt.show()

def chart3(n1, n2):
    # Chart for y = 2^x
    x = range(n1,n2)
    y = [2**n for n in x]
    plt.plot(x,y,)
    plt.xlabel('x')
    plt.ylabel('$2^{x}$')
    # plt.show()

def chart4(n1, n2):
    # Chart for y = 1/(x+1)
    x = range(n1,n2)
    y = [1/(n+1) for n in x]
    plt.plot(x,y,)
    plt.xlabel('x')
    plt.ylabel('$\\frac{1}{x+1}$')
    # plt.show()
```

```python
chart1(0,150)
```

```python
chart2(-10,11)
```

```python
chart3(-2,11)
```

```python
chart4(0,10)
```

3. Combine the figures you created in the last step into one large figure with 4 subplots

```python
plt.figure(22, (10, 8))
plt.tight_layout
plt.subplot(221)
chart1(0,150)
plt.subplot(222)
chart2(-10,11)
plt.subplot(223)
chart3(-2,11)
plt.subplot(224)
chart4(0,10)
```
