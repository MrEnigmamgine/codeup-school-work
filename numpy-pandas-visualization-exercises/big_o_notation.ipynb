Curves to graph

y = 0n + 1 and label the curve "O(1)"
y = log(n) and label the curve "O(log n)"
y = n and label the curve "O(n)"
y = n * log(n) and label it "O(n log n)"
y = n^2 and label it "O(n^2)"
y = 2^n and label it "O(2^n)"

y = n^n and label it "O(n^n)"

```python

n = np.array(range(1,11), dtype='int64')
# Curves to graph
# y = 0n + 1 and label the curve "O(1)"
y1 = 0*n +1
l1 = '$\mathcal{O}(1)$'

# y = log(n) and label the curve "O(log n)"
y2 = np.log(n)
l2 = '$0(\\log n)$'

# y = n and label the curve "O(n)"
y3 = n
l3 = '0(n)'

# y = n * log(n) and label it "O(n log n)"
y4 = n * np.log(n)
l4 = '$0({n} \\log {n})$'

# y = n^2 and label it "O(n^2)"
y5 = n**2
l5 = '$O(n^2)$'

# y = 2^n and label it "O(2^n)"
y6 = 2**n
l6 = '$O(2^n)$'

# y = n! and label it "O(n!)"
y7 = scipy.special.factorial(n)
l7 = '$O(n!)$'

# y = n^n and label it "O(n^n)"
y8 = n**n
l8 = '$O(n^n)$'
```

```python
plt.figure(1, (16,8))
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

ys = [y1, y2, y3, y4, y5, y6, y7, y8]
ls = [l1, l2, l3, l4, l5, l6, l7, l8]

# For every y
## chart (n, y$, label= l$)

for y, l in zip(ys,ls):
    plt.plot(n, y, label = l)
    # plt.show()

plt.legend()   
plt.xlim(1, 10)
plt.ylim(1, 25)
```
