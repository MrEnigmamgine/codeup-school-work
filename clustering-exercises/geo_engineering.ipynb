```python
df = wrangle.get_data()
df = df.drop_duplicates(subset=['parcelid'], keep='last')
df.latitude = df.latitude / 1_000_000
df.longitude = df.longitude / 1_000_000
# You may need to install geopy to use this code.
# %pip install geopy
geolocator = geopy.Nominatim(user_agent='zillow-project')

def get_address(lat, long, geolocator=geolocator):
    location = geolocator.reverse((lat, long))
    address =  location.raw['address']
    return address

def get_location(lat, long, geolocator=geolocator):
    location = geolocator.reverse((lat, long)).raw
    return location

def df_col_to_front(df, column):
    df = df[ [column] + [ col for col in df.columns if col != column ] ]
    return df
```
